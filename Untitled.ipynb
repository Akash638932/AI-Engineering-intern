```python
# GPA distribution
plt.figure(figsize=(7,4))
sns.histplot(df['GPA'], bins=20, kde=True)
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.show()
```

```python
# GPA by Parental Education
plt.figure(figsize=(10, 5))
sns.boxplot(x='ParentalEducation', y='GPA', data=df)
plt.title("GPA by Parental Education")
plt.xticks(rotation=45)
plt.show()
```

```python
# GPA vs Study Time
fig = plt.figure(figsize=(10, 5))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', hue='Tutoring', data=df)
plt.title("GPA vs Study Time (Tutoring highlighted)")
plt.show()
fig.savefig("gpa_by_parental_education.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Correlation Heatmap
fig = plt.figure(figsize=(10, 5))
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
fig.savefig("Correlation Heatmap.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# -----------------------------------
# Step 2: Generate Insights for a Student
# -----------------------------------
```

```python
student_id = 12345  # Replace with actual StudentID
student = df[df['StudentID'] == student_id]

if not student.empty:
    print(f"\n Performance Overview for Student {student_id}:\n")
    print(student.T)

    # Extract values
    gpa = student['GPA'].values[0]
    study_hours = student['StudyTimeWeekly'].values[0]
    absences = student['Absences'].values[0]

    # GPA Evaluation
    if gpa < 2.0:
        perf = "Weak"
    elif 2.0 <= gpa < 3.0:
        perf = "Average"
    else:
        perf = "Good"

    print(f"\n GPA: {gpa} → {perf}")
    print(f" Weekly Study Time: {study_hours} hours")
    print(f" Absences: {absences} days")

    # Call recommendation function
    generate_recommendations(student)

else:
    print(f" No student found with StudentID: {student_id}")
```

```python
# -----------------------------------
# Step 3: Recommendations
# -----------------------------------
```

```python
def generate_recommendations(student_row):
    recs = []

    if student_row['GPA'].values[0] < 3.0:
        if student_row['StudyTimeWeekly'].values[0] < 10:
            recs.append(" Increase weekly study time to at least 10–15 hours.")
        if student_row['Tutoring'].values[0] == 'No':
            recs.append(" Consider enrolling in a tutoring program.")
        if student_row['ParentalSupport'].values[0] == 'No':
            recs.append(" Seek more parental or mentor academic support.")
        if student_row['Absences'].values[0] > 5:
            recs.append(" Reduce class absences for better consistency.")
        if student_row['Extracurricular'].values[0] == 'No':
            recs.append(" Engage in extracurriculars to boost cognitive development.")

    if not recs:
        recs.append("  Keep up the good work! Focus on maintaining consistency.")

    print("\n Recommended Actions:")
    for r in recs:
        print("-", r)

if not student.empty:
    generate_recommendations(student)
```
